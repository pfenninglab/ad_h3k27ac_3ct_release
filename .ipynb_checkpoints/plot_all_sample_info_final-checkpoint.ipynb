```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rcParams
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=20
from seaborn import stripplot, violinplot
```

```python
all_h3k27ac_data = pd.read_csv("all_samples_all_extended_info.txt", sep='\t', index_col=0)
```

```python
qpcr_samples_data = pd.read_csv("qpcr_samples_all_info.txt", sep='\t')
qpcr_samples_data.index = qpcr_samples_data.proj_id
```

```python
hpc_samples_data = all_h3k27ac_data.loc[all_h3k27ac_data["brain_region"]=="HPC"]
dlpfc_samples_data = all_h3k27ac_data.loc[all_h3k27ac_data["brain_region"]=="DLPFC"]
```

```python
def plotData(data_subset, variables, dim1, dim2, title, plotTo):
    fig, axs = plt.subplots(dim1, dim2)
    fig.set_figheight(15)
    fig.set_figwidth(20)
    fig.suptitle(title)
    for i, variable in enumerate(variables):
        plotFrame = data_subset[["proj_id","binary_amyloid", variable]]
        if variable == "RSC":
            plotFrame = data_subset.loc[data_subset["NSC"]>1.01]
            plotFrame = plotFrame.loc[plotFrame["RSC"]>0.4]
            plotFrame = plotFrame.loc[plotFrame["PBC1"]>0.5]
            plotFrame = plotFrame[["proj_id","binary_amyloid", variable]]        
        else:
            plotFrame = data_subset[["proj_id","binary_amyloid", variable]]
            plotFrame = plotFrame.drop_duplicates(subset="proj_id", keep="first")

        stripplot(x="binary_amyloid",
                  y=variable,
                  hue="binary_amyloid",
                  palette={0:"blue",1:"red"},
                  size=10,
                  data = plotFrame,
                  ax=axs[i//dim2, i%dim2])
        axs[i//dim2, i%dim2].get_legend().remove()
        axs[i//dim2, i%dim2].get_xaxis().set_visible(False)
    plt.tight_layout()
    plt.savefig(plotTo)    
```

```python
variables = ["amyloid",
             "tangles",
             "cogn_global_lv",
             "cogdx",
             "braaksc",
             "ceradsc",
             "age_death",
             "educ",
             "pmi",
             "brainwt",
             "ph",
             "RSC"]
dim1 = 3
dim2 = 4
```

```python
plotData(hpc_samples_data,
         variables,
         dim1,
         dim2,
         "Hippocampus Samples",
         "hpc_samples_variables_by_abeta.svg")
```

```python
plotData(dlpfc_samples_data,
         variables,
         dim1,
         dim2,
         "DLPFC Samples",
         "dlpfc_samples_variables_by_abeta.svg")
```

```python
all_h3k27ac_data['brain_region_levels'] = all_h3k27ac_data.brain_region.map(dict(HPC=1, DLPFC=0))
all_h3k27ac_data['cell_type_levels'] = all_h3k27ac_data.cell_type.map(dict(Glia=0, Microglia=1, Neuron=2))


all_h3k27ac_data = all_h3k27ac_data.sort_values('brain_region_levels')
all_h3k27ac_data = all_h3k27ac_data.sort_values('amyloid')



quantValueColNames = ['binary_amyloid',
                      'amyloid',
                      'tangles',
                      'cogdx',
                      'brain_region_levels',
                      'cell_type_levels',
                      'age_death',
                      'msex',
                      'educ',
                      'total',
                      'filt_total',
                      'filt_unpaired_reads',
                      'mapped_pct',
                      'NRF',
                      'PBC1',
                      'NSC',
                      'RSC',
                      'FRiP',
                      'N_opt']

sampleTable = all_h3k27ac_data[quantValueColNames].values.T

scaledInfo = []
for row in sampleTable:
    cleanedRow = [x for x in row if str(x) != 'nan']
    zscores = [(val-np.min(cleanedRow))/(np.max(cleanedRow)-np.min(cleanedRow)) for val in row]
    scaledInfo.append(zscores)


sampleTableNormalized = np.array(scaledInfo)
```

```python
plt.tight_layout()
plt.figure(figsize=(20,12))
fig, ax = plt.subplots()
im = ax.imshow(sampleTableNormalized, cmap='Reds', aspect='auto')
plt.colorbar(im, orientation="horizontal")
ax.tick_params(left=False, bottom=False,
                    labelbottom=False)
ax.set_yticks(np.arange(len(quantValueColNames)))
ax.set_yticklabels([col+" (" + str(np.round(np.min(sampleTable[i,:]),1)) +"-"+ str(np.round(np.max(sampleTable[i,:]),1)) +") "   for i,col in enumerate(quantValueColNames)])

plt.savefig('all_sample_info_plotted.svg')
```

```python
qpcr_variables = ["amyloid",
             "tangles",
             "cogn_global_lv",
             "cogdx",
             "braaksc",
             "ceradsc",
             "age_death",
             "educ",
             "pmi",
             "brainwt",
             "ph"]
qpcr_dim1 = 3
qpcr_dim2 = 4
plotData(qpcr_samples_data,
         qpcr_variables,
         qpcr_dim1,
         qpcr_dim2,
         "qRT-PCR Samples",
         "qpcr_samples_variables_by_abeta.svg")
```
