```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=10
import pandas as pd
```

```python
import locale
locale.setlocale(locale.LC_ALL, 'en_US')
```

```python
ad_loci_data = pd.read_csv('ad_loci_peak_specificity.txt', sep="\t")
```

```python
kunkle_ad_loci_data = ad_loci_data.loc[ad_loci_data['GWAS study'] == "Kunkle"]
jansen_ad_loci_data = ad_loci_data.loc[ad_loci_data['GWAS study'] == "Jansen"]
eoad_loci_data = ad_loci_data.loc[ad_loci_data['Gene locus'].isin(["APP", "PSEN1", "PSEN2"])]
```

```python
ad_loci_data = ad_loci_data.sort_values(by=["GWAS study", "dominant_cell_type", "dominant_cell_type_specificity", "2nd_cell_type_specificity", "3rd_cell_type_specificity"] , ascending=[False, True,False,False,False])
```

```python
fig, ax = plt.subplots()
x_labels = [str(list(ad_loci_data["Gene locus"])[i])+" ("+str(list(ad_loci_data["Lead SNP"])[i])+", "+locale.format_string("%d", list(ad_loci_data["Closest Peak Distance"])[i], grouping=True)+"bp)" for i in range(ad_loci_data.shape[0])]
fig.set_figheight(5)
fig.set_figwidth(20)
ind = np.array([1+i*5 for i in range(ad_loci_data.shape[0])])    # the x locations for the groups
width = 1
p1 = ax.bar(ind, ad_loci_data["logFC(Neuron/other two)"], width, color="#ff9966ff", linewidth=1, edgecolor='k')
p2 = ax.bar(ind+width, ad_loci_data["logFC(Microglia/other two)"], width, color="#006699ff", linewidth=1, edgecolor='k')
p3 = ax.bar(ind+2*width, ad_loci_data["logFC(Olig/other two)"], width, color="#009966ff", linewidth=1, edgecolor='k')

ax.legend((p1[0], p2[0], p3[0]), ('Neuron', 'Microglia', "Olig"))

ax.set_title('Peak Specificity Near AD Risk Loci')
ax.set_ylabel('Cell type specificity (log2FC)')
ax.set_xticks(ind + width)
ax.tick_params(axis='x', bottom=False, labelrotation=90, labelsize=10)
ax.set_xticklabels(x_labels)

plt.savefig("ad_risk_loci_specificity.svg")
```
